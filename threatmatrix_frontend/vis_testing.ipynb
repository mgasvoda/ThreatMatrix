```python
import folium
import sqlalchemy
```

```python
def get_data(n): 
    engine = sqlalchemy.create_engine("sqlite:///C:\\Users\\600216\\sqlite\\threatmatrix.db")
    conn = engine.connect()
    metadata = sqlalchemy.MetaData(conn)
    metadata.reflect()
    events = metadata.tables['events']

    points = {'lat': [],
            'long': [],
            'fatalities': []} 

    s = sqlalchemy.select([events.c.latitude, events.c.longitude, events.c.fatalities]).limit(n)

    for row in conn.execute(s):
        points['lat'].append(row['latitude'])
        points['long'].append(row['longitude'])
        points['fatalities'].append(int(row['fatalities'])+2)

    return points
```

```python
data = get_data(250)
```

```python
m = folium.Map(location=[35, 55], zoom_start=5)
```

```python
m
```

```python
for i in range(len(data['lat'])):
    folium.Marker([float(data['lat'][i]), float(data['long'][i])], tooltip=data['fatalities'][i]).add_to(m)

m
```
